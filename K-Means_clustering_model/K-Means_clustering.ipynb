```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
print(iris.feature_names)
print(iris.data)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
mn=MinMaxScaler()
df['petal length (cm) new']=mn.fit_transform(df[['petal length (cm)']])
df['petal width (cm) new']=mn.fit_transform(df[['petal width (cm)']])
```

```python
df
```

```python
plt.scatter(df['petal length (cm) new'],df['petal width (cm) new'])
```

```python
df=df.drop(['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'],axis=1)
```

```python
df
```

```python
km=KMeans(n_clusters=3)
predicted=km.fit_predict(df)
```

```python
predicted
```

```python
df['cluster']=predicted
```
